```python
data = pd.read_csv("/content/drive/MyDrive/lab7_dataset.csv",na_values="?", header = None)
feature_list=['Hour','Immobilized_bus','Broken_Truck','Vehicle_excess','Accident_victim','Running_over','Fire_vehicles','Occurrence_involving_freight','Incident_involving_dangerous_freight','Lack_of_electricity','Fire','Point_of_flooding','Manifestations','Defect_in_the_network_of_trolleybuses','Tree_on_the_road','Semaphore_off','Intermittent_Semaphore','Slowness_in_traffic_percent']
data.columns = feature_list
data
```

```python
morning =['07:30' ,'08:00', '08:30', '09:00', '09:30','10:00' ,'10:30' ,'11:00','07:00']
noon = ['11:30','12:00' ,'12:30' ,'13:00']
afternoon =['13:30', '14:00', '14:30' ,'15:00' ,'15:30' ,'16:00']
evening =['16:30', '17:00' ,'17:30' ,'18:00']
night =['18:30' ,'19:00' ,'19:30' ,'20:00']
data['Hour'].replace(to_replace = morning,value ="Morning",inplace =True)
data['Hour'].replace(to_replace = noon,value ="Noon",inplace =True)
data['Hour'].replace(to_replace = afternoon,value ="Afternoon",inplace =True)
data['Hour'].replace(to_replace = evening,value ="Evening",inplace =True)
data['Hour'].replace(to_replace = night,value ="Night",inplace =True)
data = pd.get_dummies(data, columns = ['Hour'])
data
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
df_scaled = min_max_scaler.fit_transform(data)
df_scaled = pd.DataFrame(df_scaled)
```

```python
feature_list=['Immobilized_bus','Broken_Truck','Vehicle_excess','Accident_victim','Running_over','Fire_vehicles','Occurrence_involving_freight','Incident_involving_dangerous_freight','Lack_of_electricity','Fire','Point_of_flooding','Manifestations','Defect_in_the_network_of_trolleybuses','Tree_on_the_road','Semaphore_off','Intermittent_Semaphore','Slowness_in_traffic_percent', 'Hour_Afternoon',	'Hour_Evening',	'Hour_Morning',	'Hour_Night',	'Hour_Noon']
df_scaled.columns = feature_list
df_scaled
```

```python
X = data.drop(columns=['Slowness_in_traffic_percent'])
Y = df_scaled.iloc[:,-6: -5]
x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=0.7, shuffle=False)


X
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
mse = mean_squared_error(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)
rmse = mean_squared_error(y_test,y_pred, squared = False)
r2score = r2_score(y_test,y_pred)
try:
  confusion = confusion_matrix(y_test, y_pred)
except:
  print("Confusion Matrix Error :  This function is not supported for continuous values.")


print(f"\nMSE: {mse}")
print(f"\nMAE: {mae}")
print(f"\nRMSE: {rmse}")
print(f"\nR2 Score: {r2score}")
```
