```python
import numpy as np
import pandas as pd
from scipy.io import arff
from sklearn.preprocessing import minmax_scale
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import median_absolute_error
from sklearn.metrics import r2_score
from google.colab import drive
```

```python
drive.mount('/content/drive')
```

```python
data = arff.loadarff('/content/drive/MyDrive/Behavior of the urban traffic.arff')
data = pd.DataFrame(data[0])
```

```python
data
```

```python
data['Hour'].replace([b'7:00',b'7:30',b'8:00',b'8:30',b'9:00',b'9:30',b'10:00',b'10:30',b'11:00'],'morning',inplace=True)

data['Hour'].replace([b'11:30',b'12:00',b'12:30',b'13:00'],'noon',inplace=True)

data['Hour'].replace([b'13:30',b'14:00',b'14:30',b'15:00',b'15:30',b'16:00'],'afternoon',inplace=True)

data['Hour'].replace([b'16:30',b'17:00',b'17:30',b'18:00'],'evening',inplace=True)

data['Hour'].replace([b'18:30',b'19:00',b'19:30',b'20:00'],'night',inplace=True)
```

```python
data
```

```python
X = data.iloc[:,:-1]
```

```python
X
```

```python
Y = data.iloc[:,-1:]
```

```python
Y
```

3. Normalisation


```python
X = pd.get_dummies(X,columns = ['Hour'])
```

```python
cols=X.columns

for i in cols:
  X[i]=minmax_scale(X[i])
```

```python
X
```

# 4. Data Splitting:

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=0.8, shuffle=False)
```

# 5. Regression Models: 

## Linear Regression

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test,y_pred)
print(f"\nMSE: {mse}")
```

## Polynomial Regression

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=0.8, shuffle=False)
```

```python
poly = PolynomialFeatures(degree = 2)
```

```python
X_poly = poly.fit_transform(x_train)
```

```python
poly.fit(x_train,y_train)
```

```python
model = LinearRegression()
```

```python
model.fit(X_poly, y_train)
```

```python
y_pred = model.predict(poly.fit_transform(x_test))
```

# 5. Testing: 

```python
print(f"\nMSE: {mse}")
rmse = mean_squared_error(y_test,y_pred)
print(f"\nRMSE: {rmse}")
mea = mean_absolute_error(y_test,y_pred)
print(f"\nMEA: {mea}")
mdea = median_absolute_error(y_test,y_pred)
print(f"\nMDEA: {mdea}")
r2 = r2_score(y_test,y_pred)
print(f"\nR2 Square: {r2}")
```

## train-test ratio 70:30, 80:20, 90:10

# 7. Regression Evaluation Metrics:

```python
test = [0.3,0.2,0.1]

for i in test:
  x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=i, shuffle=False)
  poly = PolynomialFeatures(degree = 2)
  X_poly = poly.fit_transform(x_train)
  poly.fit(x_train,y_train)
  model = LinearRegression()
  model.fit(X_poly, y_train)
  y_pred = model.predict(poly.fit_transform(x_test))

  print(f"\nMSE: {mse}")
  rmse = mean_squared_error(y_test,y_pred)
  print(f"\nRMSE: {rmse}")
  mea = mean_absolute_error(y_test,y_pred)
  print(f"\nMEA: {mea}")
  mdea = median_absolute_error(y_test,y_pred)
  print(f"\nMDEA: {mdea}")
  r2 = r2_score(y_test,y_pred)
  print(f"\nR2 Square: {r2}")

  print("\n\n")
```

# 8. Playing with the Model:

## Shuffling of training samples

```python
test = [0.8,0.3,0.2,0.1]

for i in test:
  x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=i, shuffle=True)
  poly = PolynomialFeatures(degree = 2)
  X_poly = poly.fit_transform(x_train)
  poly.fit(x_train,y_train)
  model = LinearRegression()
  model.fit(X_poly, y_train)
  y_pred = model.predict(poly.fit_transform(x_test))

  print(f"\nMSE: {mse}")
  rmse = mean_squared_error(y_test,y_pred)
  print(f"\nRMSE: {rmse}")
  mea = mean_absolute_error(y_test,y_pred)
  print(f"\nMEA: {mea}")
  mdea = median_absolute_error(y_test,y_pred)
  print(f"\nMDEA: {mdea}")
  r2 = r2_score(y_test,y_pred)
  print(f"\nR2 Square: {r2}")

  print("\n\n")
```

## Degree of polynomials

```python
test = [0.8,0.3,0.2,0.1]

for i in test:
  x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=i, shuffle=False)
  poly = PolynomialFeatures(degree = 4)
  X_poly = poly.fit_transform(x_train)
  poly.fit(x_train,y_train)
  model = LinearRegression()
  model.fit(X_poly, y_train)
  y_pred = model.predict(poly.fit_transform(x_test))

  print(f"\nMSE: {mse}")
  rmse = mean_squared_error(y_test,y_pred)
  print(f"\nRMSE: {rmse}")
  mea = mean_absolute_error(y_test,y_pred)
  print(f"\nMEA: {mea}")
  mdea = median_absolute_error(y_test,y_pred)
  print(f"\nMDEA: {mdea}")
  r2 = r2_score(y_test,y_pred)
  print(f"\nR2 Square: {r2}")

  print("\n\n")
```
