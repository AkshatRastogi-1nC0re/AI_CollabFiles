```python
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/adult.data",na_values="?",names=['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','income'])
data
```

```python
#Ques.4
print("Missing values for each column:\n", data.isnull().sum())
```

```python
from sklearn.preprocessing import LabelEncoder
data_string = data.dtypes==object
data_cols = data.columns[data_string].tolist()
le = LabelEncoder()
data[data_cols] = data[data_cols].apply(lambda col: le.fit_transform(col))
data
```

```python
#Ques.7
X = data.loc[:, data.columns != 'income'].values
X
```

```python
Y = data.iloc[:,-1].values
Y
```

```python
from sklearn.naive_bayes import GaussianNB
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
gnb = GaussianNB()
y_pred = gnb.fit(x_train, y_train).predict(x_test)
```

```python
test_ConfMat_sklearn = confusion_matrix(y_test,y_pred)
print(f"Confusion Matrix on testing set: \n{test_ConfMat_sklearn}")
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:", metrics.precision_score(y_test, y_pred))
print("F1 Score:", metrics.f1_score(y_test, y_pred))
print("Recall:", metrics.recall_score(y_test, y_pred))
```
