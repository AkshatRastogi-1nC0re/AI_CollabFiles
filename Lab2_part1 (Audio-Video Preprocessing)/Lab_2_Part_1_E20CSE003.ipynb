```python
import os
for root, dirs, files in os.walk("dataset", topdown=False):
    for name in files:
        with open (os.path.join(root, name), "r") as myfile:
            data = myfile.read().splitlines()
            print(data,end="\n\n\n")
```

```python
#Remove All Special Characters

progress = []

for root, dirs, files in os.walk("dataset", topdown=False):
    for name in files:
        content = ""
        with open (os.path.join(root, name), "r") as myfile:
            data = myfile.read().splitlines()
            for i in data:
              content += i
            # content += (data*)
            xcontent = ''.join(e.lower() for e in content if (e.isalnum() and not e.isnumeric()) or e == " ")
            progress.append(xcontent)
            f = open("preprocessed/{}".format(name), "w")
            f.write(xcontent)
            f.close()
            print(xcontent, end="\n\n")
```
